```python
### LOAD RESULTS ###
import pickle

def load_Ameas(model, m=10, index=0):
    if model == "original":
        out_results = pickle.load(
            open("/home/shared/phase-retrieval/test/cGAN_Fashion/Ameas{}_cgan_4_1024_original.p".format(m), "rb")
        )
    if model == "PRCGAN":
        out_results = pickle.load(
            open("/home/shared/phase-retrieval/test/cGAN_Fashion/Ameas{}_cganrand_4_1024.p".format(m), "rb")
        )
    if model == "PRCGAN*":
        out_results = pickle.load(
            open("/home/shared/phase-retrieval/test/cGAN_Fashion/Ameas{}_cgan_4_1024.p".format(m), "rb")
        )
    if model == "DPR":
        out_results = pickle.load(
            open("/home/shared/phase-retrieval/test/VAE_Fashion/Ameas{}_3_1024.p".format(m), "rb")
        )
    if model == "E2E":
        out_results = pickle.load(
            open("/home/shared/phase-retrieval/test/FCnet_Fashion/Ameas{}_e2e_6_1024.p".format(m), "rb")
        )
    return out_results[index]
```
