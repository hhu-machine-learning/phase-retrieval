```python
### LOAD RESULTS ###
import pickle

def load_Noise(model, reg_orig=None, alpha=1.0, index=0):
    if model == "original":
        out_results = pickle.load(
            open("/home/shared/phase-retrieval/test/cGAN_MNIST/cgan_original_1024_alpha{}.p".format(alpha), "rb")
        )
    if model == "HIO":
        out_results = pickle.load(
            open("/home/shared/phase-retrieval/test/HIO_MNIST/predicted_noise_alpha{}.p".format(alpha), "rb")
        )
    if model == "E2E":
        out_results = pickle.load(
            open("/home/shared/phase-retrieval/test/FCnet_MNIST/predicted_1024_alpha{}.p".format(alpha), "rb")
        )
    if model == "PRCGAN":
        out_results = pickle.load(
            open("/home/shared/phase-retrieval/test/cGAN_MNIST/cganrand_1024_alpha{}.p".format(alpha), "rb")
        )
    if model == "PRCGAN*":
        out_results = pickle.load(
            open("/home/shared/phase-retrieval/test/cGAN_MNIST/cgan_1024_alpha{}.p".format(alpha), "rb")
        )
    if model == "DPR":
        out_results = pickle.load(
            open("/home/shared/phase-retrieval/test/VAE_MNIST/randomInit_1024_alpha{}.p".format(alpha), "rb")
        )
    if reg_orig is None:
        return out_results[index]
    else:
        return register_croco(out_results[index][None], original[None])[0]
```
